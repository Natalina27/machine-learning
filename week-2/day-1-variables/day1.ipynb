# День 1 — Переменные, числа, строки, ошибки

```python
# Что такое переменная и тип (`int` / `float` / `str` / `bool`).

# int — целое число
age = 25

# float — число с точкой
height = 1.75

# str — строка
name = "Natalya"

# bool — True/False
is_student = True

print(type(age), type(height), type(name), type(is_student))
```

```python
# Строки: конкатенация и f-строки
name = "Natalya"
age = 25

print("Меня зовут " + name + ", мне " + str(age) + " лет")
print(f"Меня зовут {name}, мне {age} лет")
```

```python
# Методы строк: .lower(), .strip()

text = "  Machine Learning  "

print(repr(text.lower()))   # без учёта регистра
print(repr(text.strip()))   # убрать пробелы по краям
print(repr(text.strip().lower()))
```

Без `let`, `const`, `var` — Python сам определяет тип по значению. Нет настоящего `const`: любую переменную можно перезаписать.

Константы — соглашение: `MAX_AGE = 100` (заглавные буквы, не защита от изменения).

```python
# Приведение типов
age = 25
text = "мне лет"

print(str(age) + " " + text)
print(f"{age} {text}")
print(int("42") + 8)
print(float("3.14") * 2)
```

```python
# Область видимости: локальная vs глобальная
name = "Natalya"

def greet():
    message = "Hello"  # локальная — только внутри функции
    print(message)
    print(name)        # глобальная — видна внутри

greet()
# print(message)  # NameError снаружи — раскомментируй и прочитай traceback
```

## Задания

```python
# 10 маленьких примеров: строки, числа, приведения типов
print(42)                         #  int
print(3.14 * 2)                   #  float
print("ML" + "-" + "101")         #  конкатенация str
print(f"score={0.81:.2f}")        #  f-строка
print(10 + 0.5)                   #  int + float
print("=" * 12)                   #  str * int
print(str(100) + "%")             #  int → str
print(int("25") + 5)              #  str → int
print(int(3.9))                   #  float → int
print(True and not False)         #  bool
```

```python
 # функция `safe_div(a, b)` (если `b=0` — вернуть `None`) 

def safe_div(a, b):
    if b == 0:
        return None
    return a / b

print(safe_div(10, 2))   # 5.0
print(safe_div(10, 0))   # None
```

```python
# функция  `is_palindrome(s)` (игнорировать пробелы и регистр).
def is_palindrome(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]

print(is_palindrome("шалаш"))  # True
print(is_palindrome("hello"))  # False
print(is_palindrome("A man a plan a canal Panama"))  # True
```

```python
# Ошибка 1 — NameError (переменная не объявлена)
print(surname)
```

```python
# NameError: surname не существует (в JS было бы ReferenceError)
# Исправление — объявить переменную:
surname = "Myunster"
print(surname)
```

```python
# Ошибка 2 — TypeError (нельзя str + int)
result = "возраст: " + 25
```

```python
# TypeError: разные типы без приведения
# Исправление:
result = "возраст: " + str(25)
print(result)
```

```python
# Ошибка 3 — IndexError (индекс вне списка)
numbers = [1, 2, 3]
print(numbers[5])
```

```python
# IndexError: индекс 5, а элементы только 0..2 (в JS было бы undefined)
# Исправление:
numbers = [1, 2, 3]
print(numbers[2])  # последний элемент
```

## Выводы

- Переменные в Python без `let` / `const` — тип выводится из значения
- Типы: `int`, `float`, `str`, `bool`; проверка: `type()`
- Строки: конкатенация, f-строки, `.lower()`, `.strip()`
- Приведение: `str()`, `int()`, `float()`
- Traceback: смотри **последнюю строку** — тип ошибки и причину
- Три частые ошибки: `NameError`, `TypeError`, `IndexError`
