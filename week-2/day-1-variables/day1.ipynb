Задание 2

```python
def safe_div(a, b):
    if b == 0:
        return None
    return a / b

print(safe_div(10, 2))   # 5.0
print(safe_div(10, 0))   # None
```

```python
def is_palindrome(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]

print(is_palindrome("шалаш"))  # True
print(is_palindrome("hello"))  # False
print(is_palindrome("A man a plan a canal Panama"))  # True
```

Задание 3

```python
# Ошибка 1 — NameError
print(surname)
```

```python
# NameError — переменная surname не была объявлена (типа ReferenceError in js)
# Исправление:
surname = "Myunster"
print(surname)
```

```python
# Ошибка 2 — TypeError
result = "возраст: " + 25
```

```python
# TypeError — нельзя складывать str и int напрямую
# Исправление:
result = "возраст: " + str(25)
print(result)
```

```python
# Ошибка 3 — IndexError
numbers = [1, 2, 3]
print(numbers[5])
```

```python
# В JS это было бы просто undefined, в Python — ошибка. 
# IndexError — индекс 5 не существует, список только из 3 элементов (0, 1, 2)
# Исправление:
numbers = [1, 2, 3]
print(numbers[2])  # последний элемент — индекс 2
```

## Выводы

- Типы: int, float, str, bool — Python определяет сам

- Три главные ошибки: NameError, TypeError, IndexError
- Приведение типов: str(), int(), float()
