```python
#dict comprehension
models = ["LinearRegression", "RandomForest", "DecisionTree"]
scores = [0.78, 0.81, 0.76]

# Dict comprehension
model_scores = {model: score for model, score in zip(models, scores)}
print(model_scores)

# С условием — только модели с accuracy > 0.77
good_models = {k: v for k, v in model_scores.items() if v > 0.77}
print(good_models)
```

```python
# Функция normalize(nums): привести список чисел в диапазон 0..1 
# (если все одинаковые — вернуть нули)
def normalize(nums):
    min_val = min(nums)
    max_val = max(nums)
    
    if min_val == max_val:
        return [0] * len(nums)
    
    return [(x - min_val) / (max_val - min_val) for x in nums]

print(normalize([10, 20, 30, 40, 50]))
print(normalize([5, 5, 5]))  # все одинаковые
```

```python
# то же самое через обычный цикл (без comprehension)
def normalize_loop(nums):
    min_val = min(nums)
    max_val = max(nums)
    
    if min_val == max_val:
        return [0] * len(nums)
    
    result = []
    for x in nums:
        result.append((x - min_val) / (max_val - min_val))
    return result

print(normalize_loop([10, 20, 30, 40, 50]))
print(normalize_loop([5, 5, 5]))  # все одинаковые
```

```python
# Собрать мини‑пайплайн: фильтр → преобразование → агрегация (всё через функции)
# Pipeline -> sklearn — фильтр → трансформация → агрегация

def filter_scores(scores, threshold=0.5):
    return [x for x in scores if x >= threshold]

def transform_scores(scores):
    return normalize(scores)

def aggregate(scores):
    return sum(scores) / len(scores) if scores else 0

# Пайплайн
raw_scores = [0.3, 0.7, 0.8, 0.2, 0.9, 0.6]

filtered = filter_scores(raw_scores)
transformed = transform_scores(filtered)
result = aggregate(transformed)

print(f"После фильтра: {filtered}")
print(f"После нормализации: {transformed}")
print(f"Среднее: {result:.2f}")
```
