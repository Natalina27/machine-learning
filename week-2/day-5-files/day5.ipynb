## Работа с файлами
### open/read/write, кодировка UTF‑8
**Режимы открытия:**
- `open("file.txt", "r")` — открыть для чтения
- `open("file.txt", "w")` — открыть для записи (перезаписывает!)
- `open("file.txt", "a")` — добавить в конец
- `with open(...) as f:` — автоматически закрывает файл
- `encoding="utf-8"` — поддержка всех языков включая русский
**Методы чтения:**
- `f.read()` — весь файл как одна строка
- `f.readlines()` — список строк
- `f.readline()` — одна строка
**Методы записи:**
- `f.write("текст")` — записать строку
- `f.writelines(["строка1\n", "строка2\n"])` — записать список строк

### pathlib — работа с путями
- `Path(".")` — текущая папка
- `Path("data") / "file.csv"` — объединить пути (вместо строк)
- `path.exists()` — проверить существование
- `path.mkdir(exist_ok=True)` — создать папку
- `path.iterdir()` — список файлов в папке
- Преимущество: работает на всех ОС (не нужно думать о `/` или `\`)

```python
from pathlib import Path
# Текущая папка
current = Path(".")
print(f"Текущая папка: {current.absolute()}")
# Создать путь
file_path = Path(".") / "sample.txt"
print(f"Путь к файлу: {file_path}")
print(f"Файл существует: {file_path.exists()}")
# Список файлов в папке
print("\nФайлы в папке:")
for f in Path(".").iterdir():
    print(f" {f.name}")
```

## Задания

```python
# Взять текстовый файл (лог/статья) и посчитать: 
# количество строк, уникальные слова, топ‑10 слов
from collections import Counter
with open("sample.txt", "r", encoding="utf-8") as f:
    words = f.read().split()
# Топ-10 слов
top_words = Counter(words).most_common(10)
print("Топ-10 слов:")
for word, count in top_words:
    print(f"  {word}: {count}")
unique_words = set(words)
print(f"Уникальных слов: {len(unique_words)}")
# количество строк
with open("sample.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
print(f"Количе��тво строк: {len(lines)}")
```

```python
# Сохраняем результат в файл
with open("results.txt", "w", encoding="utf-8") as f:
    f.write(f"Количество строк: {len(lines)}\n")
    f.write(f"Уникальных слов: {len(unique_words)}\n\n")
    f.write("Топ-10 слов:\n")
    for word, count in top_words:
        f.write(f"{word}: {count}\n")
print("Результаты сохранены в results.txt!")
```
