## Цель
Уверенно работать со структурами данных Python.

```python
# List — упорядоченная коллекция (типа array)
scores = [85, 92, 78, 95, 88]
scores1 = [1, 2, 3, 4, 5, 10, 11]
print(scores[0])      # первый элемент
print(scores[-1])     # последний элемент
print(scores[1:3])    # срез — элементы 1 и 2
print(len(scores))    # длина списка
```

```python
scores.append(91)        # добавить в конец
scores.extend([70, 80])  # добавить несколько
scores.sort()            # сортировка
scores1.sort()
print(scores)
print(scores1)
```

```python
# Dict — ключ: значение (типа object)
model_scores = {
    "LogisticRegression": 0.80,
    "RandomForest": 0.81,
    "DecisionTree": 0.76
}
print(model_scores["RandomForest"])        # получить значение
print(model_scores.get("SVM", "не найдено"))  # безопасный доступ
print(model_scores.get("SVM"))  # curiosity :)) - None
print(model_scores.keys())                 # все ключи
print(model_scores.values())              # все значения
```

```python
# Set — уникальные значения
features_1 = {"age", "income", "sex"}
features_2 = {"age", "fare", "sex"}
print(features_1 & features_2)  # пересечение
print(features_1 | features_2)  # объединение
print(features_1 - features_2)  # разница
```

```python
# главное свойство set — только уникальные значения:
duplicates = ["age", "age", "income", "sex", "income"]
unique = set(duplicates)
print(unique)
```

```python
# копирование списков
# Опасно! (ссылка на один и тот же объект)
a = [1, 2, 3]
b = a
b.append(4)
print(a)  # [1, 2, 3, 4]
print(b)  # [1, 2, 3, 4]
```

```python
# исправить — используй .copy():
pythona = [1, 2, 3]
b = a.copy()
b.append(4)
print(a)  # [1, 2, 3] — не изменился
print(b)  # [1, 2, 3, 4]
```

## Задания

```python
# словарь частот и вывести топ‑3 самых частых.
words = ["ML", "Python", "ML", "pandas", "Python", "ML", "numpy"]
frequency = {}
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1
print(frequency)
print(list(frequency.keys())[:3])
```

```python
# Функция top_k(scores: dict, k) → список (ключ, значение) по убыванию
def top_k(scores: dict, k):
    return sorted(scores.items(), key=lambda x: x[1], reverse=True)[:k]
print( top_k(frequency, 3))
print(top_k(frequency, len(frequency)))
```

```python
# Мини‑демо: вложенный список, где изменение "течёт" (пояснить почему).
matrix = [[1, 2], [3, 4]]
matrix_copy = matrix.copy() # shallow copy — копирует список, но вложенные списки остаются ссылками.
matrix_copy[0].append(99)
print(matrix)       # [[1, 2, 99], [3, 4]]
print(matrix_copy) # [[1, 2, 99], [3, 4]]
```

```python
# исправить — нужен deep copy:
import copy
matrix = [[1, 2], [3, 4]]
matrix_deep = copy.deepcopy(matrix)
matrix_deep[0].append(99)
print(matrix)       # [[1, 2], [3, 4]] — не изменился
print(matrix_deep)  # [[1, 2, 99], [3, 4]]
```

## Выводы
- **list** — упорядоченная коллекция, индексы, срезы, `[-1]` = по��ледний элемент
- `sort()` в Python умный — числа сортирует как числа (в JS нужен колбек)
- **dict** — ключ:значение; `.get()` безопаснее чем `[]` (нет KeyError)
- `lambda x: x[1]` — анонимная функция, как callback function в JS
- **set** — только уникальные значения; операции: `&`, `|`, `-`
- `b = a` — ссылка, не копия! Используй `b = a.copy()`
- `b = a.copy()` — shallow copy: вложенные объекты всё ещё ссылки
- `copy.deepcopy(a)` — deep copy: полностью независимая копия
