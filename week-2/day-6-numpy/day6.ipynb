## Цель
Понять зачем NumPy нужен и как делать вычисления без циклов.

````python
# почему NumPy быстрее списка - потому что NumPy написан на C 
# и работает с памятью напрямую — без Python overhead на каждый элемент.
import numpy as np
import time
# Обычный список
big_list = list(range(1_000_000))
start = time.time()
result = [x * 2 for x in big_list]
print(f"Список: {time.time() - start:.4f} сек")
# NumPy массив
big_array = np.arange(1_000_000)
start = time.time()
result = big_array * 2
print(f"NumPy: {time.time() - start:.4f} сек")
````

## ndarray vs list
| | Python list | NumPy ndarray |
|---|---|---|
| Типы элементов | разные (`[1, "a", True]`) | один тип (`int64`, `float64`) |
| Скорость | медленнее | быстрее (C под капотом) |
| Математика | нужен цикл | работает сразу на всём массиве |
| Память | больше | меньше |
## dtype — тип данных всех элементов
- `int64` — целые числа
- `float64` — числа с точкой
- `bool` — True/False
## shape и reshape
- `shape` — размерность массива: `(3,)` для вектора, `(2, 3)` для матрицы
- `reshape` — изменить форму без изменения данных:
```python
arr = np.arange(6)         # [0, 1, 2, 3, 4, 5]
matrix = arr.reshape(2, 3) # [[0,1,2],[3,4,5]]
```

````python
# ndarray — основной тип NumPy
arr = np.array([1, 2, 3, 4, 5])
print(f"Тип: {type(arr)}")
print(f"dtype: {arr.dtype}")      # тип данных внутри
print(f"shape: {arr.shape}")      # размерность
print(f"ndim: {arr.ndim}")        # количество измерений
````

````python
# shape — размерность массива:
arr = np.array([1, 2, 3])        # shape (3,)
matrix = np.array([[1,2],[3,4]]) # shape (2, 2)
````

````python
# reshape — изменить форму без изменения данных:
arr = np.arange(6)        # [0, 1, 2, 3, 4, 5]
matrix = arr.reshape(2, 3) # [[0,1,2],[3,4,5]]
````

````python
# Создаём массив и меняем форму
arr = np.arange(6)
print(f"До reshape: {arr} shape: {arr.shape}")
matrix = arr.reshape(2, 3)
print(f"После reshape:\n{matrix}")
print(f"shape: {matrix.shape}")
````

````python
# индексация и булевы маски
arr = np.array([10, 20, 30, 40, 50])
# Обычная индексация
print(arr[0])      # первый элемент
print(arr[-1])     # последний
print(arr[1:3])    # срез
# Булева маска — выбрать элементы по условию
mask = arr > 25
print(f"Маска: {mask}")
print(f"Элементы > 25: {arr[mask]}")
````

## mean/std по оси
- `arr.mean()` — среднее всего массива
- `arr.mean(axis=0)` — среднее по строкам (вдоль колонок)
- `arr.mean(axis=1)` — среднее по колонкам (вдоль строк)
- `arr.std()` — стандартное отклонение

````python
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
print(f"Среднее всего: {matrix.mean()}")
print(f"Среднее по оси 0 (по строкам): {matrix.mean(axis=0)}")
print(f"Среднее по оси 1 (по колонкам): {matrix.mean(axis=1)}")
````

````python
# broadcasting — одна из самых важных концепций NumPy
# Без broadcasting — нужен цикл
arr = np.array([1, 2, 3, 4, 5])
# С broadcasting — Python сам "растягивает" число
result = arr + 10
print(result)  # [11, 12, 13, 14, 15]
# Ещё пример — матрица + вектор
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
row = np.array([10, 20, 30])
print(matrix + row)  # row "растягивается" на каждую строку
````

## Задания

````python
# Сгенерировать массив роста/веса, посчитать BMI и 
# отфильтровать значения выше порога
# Генерируем данные роста и веса
np.random.seed(42)
height = np.random.uniform(1.5, 2.0, 100)  # рост в метрах
weight = np.random.uniform(50, 100, 100)   # вес в кг
# Считаем BMI
bmi = weight / height ** 2
print(f"Средний BMI: {bmi.mean():.2f}")
print(f"Мин BMI: {bmi.min():.2f}")
print(f"Макс BMI: {bmi.max():.2f}")
# Фильтруем — BMI выше 25 (избыточный вес)
high_bmi = bmi[bmi > 25]
print(f"Людей с BMI > 25: {len(high_bmi)} из {len(bmi)}")
````

````python
# Функция standardize(x): z‑score со страховкой от нулевой дисперсии.
def standardize(x):
    """Z-score нормализация со страховкой от нулевой дисперсии"""
    std = x.std()
    if std == 0:
        return np.zeros_like(x)
    return (x - x.mean()) / std
arr = np.array([10, 20, 30, 40, 50])
print(standardize(arr))
# Страховка
arr_same = np.array([5, 5, 5, 5])
print(standardize(arr_same))
````

````python
# Сравнить скорость (на маленьком примере): цикл по списку vs операция NumP
import time
# Генерируем большой массив
size = 10_000_000
# Цикл по списку
py_list = list(range(size))
start = time.time()
result_list = [x ** 2 for x in py_list]
list_time = time.time() - start
# NumPy операция
np_array = np.arange(size)
start = time.time()
result_np = np_array ** 2
np_time = time.time() - start
print(f"Список: {list_time:.4f} сек")
print(f"NumPy:  {np_time:.4f} сек")
print(f"NumPy быстрее в {list_time/np_time:.1f} раз")
````
