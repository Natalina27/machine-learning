```python
# мини-калькулятор
def calculator(a, op, b):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        if b == 0:
            return None
        return a / b
    else:
        return None

print(calculator(10, "+", 5))   # 15
print(calculator(10, "/", 0))   # None
print(calculator(10, "/", 2))   # 5.0
```

```python
# счётчик — по строке посчитать количество букв, цифр и пробелов
alpha = 0
digit = 0
space = 0

text = "Hello, world! 123"
for char in text:
    if char.isalpha():
        alpha += 1
    elif char.isdigit():
        digit += 1
    elif char.isspace():        
        space += 1

print(f"Букв: {alpha}, цифр: {digit}, пробелов: {space}")
```

```python
# найти максимум и его индекс в списке без использования max()
numbers = [1, 2, 3, 4, 5, 9, 22, 66, 7]
max_value = numbers[0]
max_index = 0

for i in range(1, len(numbers)):
    if numbers[i] > max_value:
        max_value = numbers[i]
        max_index = i   
print(f"Максимальное значение: {max_value}, его индекс: {max_index}")
```
