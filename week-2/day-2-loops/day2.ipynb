```python
# if/elif/else 
accuracy = 0.85
if accuracy >= 0.80:
    print("Модель хорошая")
elif accuracy >= 0.60:
    print("Модель средняя, нужно улучшить")
else:
    print("Модель плохая, переобучаем")
```

```python
# логика (and/or/not)
accuracy = 0.75
has_data = True
# and — все условия должны быть True
if accuracy >= 0.60 and has_data:
    print("Можно обучать модель")
# or — достаточно одного True (простой пример)
score_weak = 0.45
score_strong = 0.92
if score_weak < 0.50 or score_strong > 0.90:
    print("Хотя бы одно значение выходит за порог")
# or на одной переменной
if accuracy < 0.50 or accuracy > 0.90:
    print("accuracy на краю диапазона")  # при 0.75 не сработает
# or + not (частый случай на практике)
if accuracy < 0.60 or not has_data:
    print("Нужны данные или лучше качество")
# not — инверсия
is_overfitting = False
if not is_overfitting:
    print("Переобучения нет")
```

```python
# for по списку
models = ["LinearRegression", "LogisticRegression", "RandomForest"]
for model in models:
    print(f"Обучаю модель: {model}")
```

```python
# for по строке — строка тоже итерируемая (посимвольно)
model_name = "RandomForest"
for char in model_name:
    print(char, end=" ")
print()  # перенос строки в конце
# подсчёт гласных в названии
vowels = "aeiouAEIOU"
count = 0
for char in model_name:
    if char in vowels:
        count += 1
print(f"Гласных в '{model_name}': {count}")
```

```python
# for по диапазону range
for i in range(5):
    print(f"Эпоха {i+1} из 5")
```

```python
# while — когда не знаем заранее сколько итераций
count = 0
while count < 5:
    print(f"Итерация {count}")
    count += 1
```

## Задания

```python
# мини-калькулятор
def calculator(a, op, b):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        if b == 0:
            return None
        return a / b
    else:
        return None
print(calculator(10, "+", 5))   # 15
print(calculator(10, "/", 0))   # None
print(calculator(10, "/", 2))   # 5.0
```

```python
# счётчик — по строке посчитать количество букв, цифр и пробелов
alpha = 0
digit = 0
space = 0
text = "Hello, world! 123"
for char in text:
    if char.isalpha():
        alpha += 1
    elif char.isdigit():
        digit += 1
    elif char.isspace():
        space += 1
print(f"Букв: {alpha}, цифр: {digit}, пробелов: {space}")
```

```python
# найти максимум и его индекс в списке без использования max()
numbers = [1, 2, 3, 4, 5, 9, 22, 66, 7]
max_value = numbers[0]
max_index = 0
for i in range(1, len(numbers)):
    if numbers[i] > max_value:
        max_value = numbers[i]
        max_index = i
print(f"Максимальное значение: {max_value}, его индекс: {max_index}")
```

## Выводы
- `if` / `elif` / `else` — без скобок, с двоеточием и отступами
- `elif` вместо `else if` как в JS
- `and` / `or` / `not` — логические операторы для составных условий
- `for item in list` — проход по элементам списка
- `for char in str` — проход по символам строки
- `for i in range(n)` — счётчик от 0 до n-1
- `while` — цикл, пока условие True (осторожно с бесконечным циклом)
