# День 1 — Git/GitHub и структура проекта

## Цель
Создать рабочую структуру week-3 и настроить Git workflow: каждый день — осмысленный commit и push на GitHub.

## Git — основные понятия

- **Repository** — папка проекта + вся история изменений.
- **Commit** — снимок файлов с сообщением (*что изменилось?*).
- **Branch** — отдельная линия работы (например `week-3`).
- **Remote** — копия репозитория на GitHub.
- **Push** — отправить локальные коммиты на remote.

История коммитов — часть результата обучения, не техническая мелочь.

## Файлы проекта

| Файл | Зачем |
|------|-------|
| `README.md` | цель, структура, как запустить |
| `requirements.txt` | зафиксированные библиотеки |
| `.gitignore` | что **не** попадает в Git |

**Не коммитим:** `.venv`, `__pycache__`, `.ipynb_checkpoints`, `.env`, токены, большие датасеты.

Материалы: [Pro Git (рус.)](https://git-scm.com/book/ru/v2) · [GitHub Hello World](https://docs.github.com/en/get-started/start-your-journey/hello-world) · [Ignoring files](https://docs.github.com/en/get-started/git-basics/ignoring-files)

Запускай ноутбук из **корня репозитория** `machine-learning`.

````python
import subprocess
from pathlib import Path

def run(cmd: str) -> None:
    print(f"$ {cmd}")
    result = subprocess.run(cmd, shell=True, capture_output=True, text=True)
    out = (result.stdout or result.stderr).strip()
    if out:
        print(out)
    print()

# Корень репо: из machine-learning/ или из week-3/notebooks/
REPO = Path(".")
if not (REPO / "week-3").exists():
    REPO = Path("../..")

print(f"Корень репо: {REPO.resolve()}")
````

````python
# Проверка Git: версия, имя, email, ветка, remote
run("git --version")
run("git config user.name")
run("git config user.email")
run("git branch --show-current")
run("git remote -v")
````

Если `user.name` / `user.email` пустые:

```bash
git config --global user.name "Твоё Имя"
git config --global user.email "you@example.com"
```

````python
# Проверка структуры week-3 и ключевых файлов в корне
WEEK3 = REPO / "week-3"
expected = [
    REPO / "README.md",
    REPO / "requirements.txt",
    REPO / ".gitignore",
    WEEK3 / "notebooks",
    WEEK3 / "notebooks" / "day01_git_setup.ipynb",
]

for p in expected:
    mark = "OK" if p.exists() else "MISSING"
    print(f"  [{mark}] {p.relative_to(REPO)}")
````

## Задания

````python
# 1. git --version работает; user.name и user.email настроены
git_name = subprocess.run("git config user.name", shell=True, capture_output=True, text=True).stdout.strip()
git_email = subprocess.run("git config user.email", shell=True, capture_output=True, text=True).stdout.strip()
version = subprocess.run("git --version", shell=True, capture_output=True, text=True).stdout.strip()

print(version)
print(f"user.name:  {git_name or '(не задан — настрой git config)'}")
print(f"user.email: {git_email or '(не задан — настрой git config)'}")
````

````python
# 2. Remote на GitHub настроен (origin → твой репозиторий)
remote = subprocess.run("git remote -v", shell=True, capture_output=True, text=True).stdout.strip()
print(remote if remote else "remote не настроен — добавь: git remote add origin <url>")
````

````python
# 3. Структура week-3/notebooks/ на месте
nb_dir = REPO / "week-3" / "notebooks"
notebooks = sorted(nb_dir.glob("day*.ipynb")) if nb_dir.exists() else []

print(f"Ноутбуков в week-3/notebooks/: {len(notebooks)}")
for nb in notebooks:
    print(f"  {nb.name}")
````

````python
# 4. README.md, requirements.txt, .gitignore — есть и не пустые
for filename in ["README.md", "requirements.txt", ".gitignore"]:
    path = REPO / filename
    size = path.stat().st_size if path.exists() else 0
    print(f"{filename}: {size} bytes")
    if size:
        print(path.read_text(encoding="utf-8")[:300])
        print("...\n" if size > 300 else "\n")
````

````python
# 5. git status — что изменилось перед commit/push
run("git status -sb")
````

- `README.md`, `requirements.txt`, `.gitignore` — обязательные файлы в корне.
- Ноутбуки week-3 лежат в `week-3/notebooks/`.
- Перед push: **Kernel → Restart & Run All** — не отправляй битый ноутбук.
